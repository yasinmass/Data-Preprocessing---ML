## Frequency encoding 

*description : Frequency Encoding replaces each category in a feature with the number of times it appears in the dataset.*

```python
import numpy as np 
import pandas as pd 
```

dataset - benz.csv

```python
benz=pd.read_csv("dataset/benz.csv" , usecols={'X1','X2'})
benz
```

```python
#count the col categories
for col in benz.columns[0:]:
    print(f"{col} : {len(benz[col].value_counts())}")
    
```

```python
#X1 frequency in dict from
count_values=benz['X1'].value_counts().to_dict()
count_values
```

```python
#map the dict to X1
benz["X1"]=benz["X1"].map(count_values)
benz.head()
```


# When to Use

- When dataset has high cardinality (many categories)

- When you want to avoid too many columns

- Useful in tree-based models like Random Forest, XGBoost

**disadvantages**
- Different categories may get same frequency

- Does not capture category meaning

- May introduce unintended importance( higher frequency gets more important )
