## leave out one encoding

*For each row, replace the category with the average of the target values of the same category — but don’t include that row’s own target value in the average.

```python
import pandas as pd 
import category_encoders as ce
```

```python
data=pd.DataFrame({'name':['A','A','c','D','E'],'mark':[100,90,80,70,60]})
data
```

```python
# Create encoder
encoder = ce.LeaveOneOutEncoder(cols=['name'])

# Fit and transform
encoded = encoder.fit_transform(data['name'], data['mark'])
print(encoded)
```

# Pros

Reduces overfitting compared to Target Encoding

Handles high-cardinality features

Works well in supervised ML

# Cons

Still risk of leakage if not used properly

Computationally heavier than Label/One-Hot

Needs enough data per category

When to use

| Situation           | Use LOO?  |
| ------------------- | --------- |
| Supervised learning | ✅ Yes     |
| High-cardinality    | ✅ Yes     |
| Small dataset       | ⚠ Careful |
| Unsupervised        | ❌ No      |
