## label encoding - to map data with catagorical data

```python
import pandas as pd 
```

creating a sample custumer data

```python
custumer_data=pd.DataFrame({"custumer_id" : [1,2,3,4,5] , "satisfaction":["low","high","medium","high","low"]})
custumer_data
```

*mapping low < medium < high*

```python
data_map={"low":0,"medium":1,"high":2}
```

```python
custumer_data["satisfaction"]=custumer_data["satisfaction"].map(data_map)
custumer_data
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
custumer_data['Encoded_sklearn'] = le.fit_transform(custumer_data['satisfaction'])

print(custumer_data)
```
